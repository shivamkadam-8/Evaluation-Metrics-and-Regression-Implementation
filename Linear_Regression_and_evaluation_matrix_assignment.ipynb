2) **What are the assumption of linear regression?**

**Linearity**

The relationship between Independent variable and dependent variable is linear.This means that the model assumes the target variable can be expressed a linear combination the predictor.
If the true relationship is nonlinear the linear regression is might not provide an accurate model.

**Independence of Errors**

The errors (residuals) ‚Äî the differences between the observed and predicted values ‚Äî should be independent of each other. This means that the residual at one point should not provide any information about the residual at another point.

**Homoscedasticity**

The residuals should have constant variance across all levels of the independent variables. In other words, the spread or "scatter" of residuals should be roughly the same for all predicted values.
If the variance of the errors increases or decreases with the level of the predicted values (a phenomenon called heteroscedasticity), it can make the model's estimates less reliable.

**Normality of Errors**

The residuals (errors) should be approximately normally distributed. This assumption is especially important for inference, such as hypothesis testing or constructing confidence intervals for the regression coefficients.
If the errors are not normally distributed, it may affect the validity of statistical tests, but the model's predictions might still be accurate.

**No Multicollinearity**

The independent variables should not be highly correlated with each other. High correlation between predictors (multicollinearity) makes it difficult to estimate the relationship between each individual predictor and the target variable.
When multicollinearity is present, it can lead to inflated standard errors and unstable coefficient estimates.

**No Endogeneity**

The independent variables should not be correlated with the error term. In other words, there should be no omitted variable bias (i.e., no important variables left out of the model that are correlated with both the predictors and the outcome).
If endogeneity is present, the regression estimates become biased and inconsistent.

**No Outliers or Influential Data Points**

While outliers and influential data points are not formal assumptions of linear regression, their presence can severely affect the performance and validity of a linear regression model. It's essential to check for and address them, as they can violate the underlying assumptions of the model and lead to misleading results.

Here‚Äôs how outliers and influential data points interact with linear regression:

**Outliers in Linear Regression**

Definition: Outliers are data points that deviate significantly from the overall pattern or trend in the data. They can arise from errors in data collection or represent rare events.
Impact: Outliers can distort the results of linear regression, especially by skewing the regression coefficients or increasing the error variance.

**Influential Data Points in Linear Regression**

**Definition**: Influential data points are those that have a large impact on the estimated regression coefficients or on the overall fit of the model. These points may not be extreme in terms of their values (like outliers), but their position can disproportionately influence the model.
Impact: Even if a data point is not an outlier, if it is far from the mean of the predictor variables, it can skew the regression results.

R-squared and Adjusted R-squared  are both statistical measures used to evaluate the goodness of fit of a regression model,

**R-squared**

R-squared is a measure of the proportion of the variance in the dependent variable that is explained by the independent variables in the model. It is often called the "coefficient of determination."

**Adjusted R-squared**

Definition: Adjusted R-squared adjusts R¬≤ for the number of predictors in the model. It provides a more accurate measure of goodness of fit, particularly when comparing models with different numbers of predictors.

Range: Like R¬≤, Adjusted R¬≤ is between 0 and 1, but it can also become negative if the model is a poor fit.

If Adjusted R¬≤ increases when adding a predictor, it means the new predictor improves the model's fit.

R¬≤ always increases (or stays the same) as more predictors are added, even if they don't improve the model‚Äôs performance.

7) **What is multicolinearity , and how does it impact regression?**

Multicollinearity refers to a situation in a regression model where two or more independent variables (predictors) are highly correlated with each other. This can cause issues in estimating the coefficients of the regression model and interpreting the results accurately.

**Unstable Coefficients**:

When independent variables are highly correlated, the model has difficulty in distinguishing the individual effect of each predictor on the dependent variable.

**Reduced Interpretability**:

The presence of multicollinearity makes it difficult to interpret the effect of individual predictors, as the model may not be able to clearly separate their individual contributions to the outcome variable.

**Overfitting**:

Multicollinearity can lead to overfitting, where the model becomes too complex and starts capturing noise rather than the true underlying relationship between the predictors and the dependent variable.

8) **What is Mean Absolute error ?**

Mean Absolute Error (MAE) is a metric used to measure the accuracy of a regression model. It represents the average of the absolute differences between the predicted values and the actual values in the dataset.

MAE provides an average of the absolute errors, meaning it gives you a straightforward sense of how far off your predictions are from the actual values, without considering whether the errors are positive or negative (which is the case with MSE or RMSE).

Since the errors are not squared (like MSE), MAE is in the same units as the target variable, making it easier to interpret in real-world terms. For example, if your target variable represents prices in dollars, the MAE will also be in dollars, which is intuitive.

MAE is less sensitive to outliers than metrics like Mean Squared Error (MSE), because it doesn't square the errors. Large errors don't have as much impact on the final MAE as they would on MSE.

**Robustness to Outliers**: Unlike MSE, MAE doesn't amplify the impact of large outliers because it uses absolute values instead of squaring the differences.

**Non-differentiable**:MAE is not differentiable at zero (because of the absolute value), which can make it more challenging to optimize during the training of models in some machine learning algorithms, like gradient-based optimization.

**Less Sensitive to Large Errors**

Since it doesn't square the errors, MAE doesn't penalize large errors as much as MSE does. This means that MAE might not be as effective in situations where you want to put more weight on reducing larger errors.

MAE is useful when you want a simple, interpretable metric that gives you the average magnitude of error without giving too much weight to larger errors.

10) **Why is RMSE considered more interpretable than MSE?**

Root Mean Squared Error (RMSE) is often considered more interpretable than Mean Squared Error (MSE) because it has the same units as the target variable being measured.

**Units of Measurement:**

**MSE** is the average of squared differences between the predicted and actual values. Since it involves squaring the error, the result is in squared units. For example, if you're predicting temperatures in Celsius, MSE would be in square Celsius (¬∞C¬≤), which can be hard to interpret.

**RMSE**, on the other hand, takes the square root of MSE, which brings it back to the original units of the target variable (e.g., Celsius, meters, dollars, etc.). This makes RMSE much easier to understand in the context of the data, as it represents the typical error in the same units as the data itself.

**Practical Interpretation:**

With RMSE, you can directly interpret the magnitude of the error in terms of the scale of the data. For example, if RMSE is 5¬∞C when predicting temperatures, it means that the typical error in the predictions is around 5¬∞C.

MSE, because it uses squared units, doesn't have as straightforward an interpretation in a real-world context.

**Scale Sensitivity**:

Both MSE and RMSE are sensitive to large errors, but RMSE provides a more interpretable measure of the average magnitude of those errors.

14) **How can we address multicolinearity in regression?**

Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other. This can lead to problems such as inflated standard errors, unreliable coefficient estimates, and difficulty in determining the individual effect of each predictor. There are several methods to address multicollinearity in regression:

**Remove Highly Correlated Variables**

If two or more variables are highly correlated, consider removing one of them from the model. This is often the simplest and most effective solution.

**Principal Component Analysis (PCA)**

PCA can be used to reduce the dimensionality of the dataset by transforming correlated variables into a smaller set of uncorrelated components. This can help eliminate multicollinearity, but the resulting components might not have a straightforward interpretation.

**Ridge Regression**

Ridge regression adds a penalty term to the regression equation that discourages large coefficients. This helps to reduce the impact of multicollinearity by shrinking the coefficients of highly correlated predictors. The result is a more stable model.

16) **How is adjusted R-squared calculated?**

Adjusted R-squared is a modified version of the standard R-squared (coefficient of determination) that accounts for the number of predictors (independent variables) in a regression model. Unlike R-squared, which always increases when you add more variables to the model, adjusted R-squared penalizes the addition of irrelevant predictors that don't improve the model significantly.

**R-squared**: This is the proportion of variance in the dependent variable that is explained by the independent variables in the model. However, R-squared tends to increase when you add more predictors, even if they are not meaningful, which can be misleading.

**Adjustment for the Number of Predictors**: The formula for adjusted R-squared incorporates the number of predictors (p) and the number of observations (n) to adjust the value of R-squared. This adjustment ensures that adding more predictors only increases the adjusted R-squared if the new variables improve the model in a meaningful way.

If adding a predictor improves the model (i.e., increases the explanatory power of the model), the adjusted R-squared will increase.
If the predictor does not significantly improve the model, the adjusted R-squared will decrease.

19) **What is Root Mean Squared error(RMSE)?**

Root Mean Squared Error (RMSE) is a commonly used metric to evaluate the performance of a regression model. It measures the average magnitude of the residuals (errors) between the predicted values and the actual values, with a focus on penalizing larger errors more heavily due to the squaring of the differences.

**Interpretability**:

RMSE is in the same unit as the dependent variable, which makes it easier to interpret compared to some other metrics (like R-squared, which is a unitless metric).
A lower RMSE indicates better model performance (smaller errors), while a higher RMSE indicates that the model's predictions are farther from the actual values.

**Sensitivity to Large Errors**:

Since RMSE squares the differences between actual and predicted values, it is particularly sensitive to larger errors (outliers). A large error will have a disproportionately large effect on RMSE, making it a more stringent measure of model performance.
This is both an advantage (since it penalizes large errors) and a disadvantage (since it may overemphasize outliers).

**Average Magnitude of Error:**

RMSE represents the square root of the average squared error, which provides a measure of how spread out the errors are around the regression line. It gives an idea of how much the predicted values deviate from the actual values on average.

20) **Why is pickling considered risky?**

**Security Risks (Arbitrary Code Execution):**

The most significant risk with pickling is that it can execute arbitrary code during the deserialization (unpickling) process. If you unpickle data from an untrusted or unknown source, the pickle file could contain malicious code.
When you unpickle an object, Python will reconstruct the object by calling its __reduce__ method, which can execute arbitrary code if the data is crafted to exploit this. This makes it possible for an attacker to inject harmful code that could damage your system or steal sensitive information.

**Compatibility Issues**:

Pickle is Python-specific and doesn't ensure backward or forward compatibility. This means that pickled data may not load correctly if you change your Python version, or if the Python environment has different libraries or object definitions than when the data was originally pickled.
If you try to unpickle data that was serialized with a different version of Python or with changes to class structures, you might encounter errors or the data might not deserialize correctly.

**Untrusted Sources**:

If you are dealing with data from an untrusted source, unpickling it can expose your system to arbitrary code execution, which is why it is generally not safe to unpickle data from sources you don't control or trust (like data from a web server or a third-party API).
This is especially important if you're using pickling to store user-generated data, as an attacker could send you a malicious pickle object to execute harmful actions.

Identifying Synergies or Conflicts: Sometimes, the combination of variables may have a stronger or weaker effect than the individual effects of each variable. Interaction terms can help identify these synergies or conflicts. For example, in a marketing campaign, the effect of a discount might depend on whether the consumer is a new or returning customer.
